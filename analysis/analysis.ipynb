```python
def plot_conf(conf):
    fig = plt.figure(figsize = (20,14))
    sn.heatmap(conf, annot=False, xticklabels = class_names,  yticklabels = class_names)
    plt.ylabel('True')
    plt.xlabel('Predicted')
    fig.savefig('plot.pdf')
    plt.show()

def plot_hist(hist):
    fig = plt.figure()
    hist = hist.item()
    train = hist['loss']
    val   = hist['val_loss']
    plt.plot(train)
    plt.plot(val)
    plt.title('Loss')
    plt.ylabel('loss')
    plt.xlabel('epoch')
    plt.legend(['Training', 'Validation'], loc='upper right')
    plt.show()
    fig.savefig('plot.pdf')

def plot_fine_hist(hist1, hist2, hist3):
    plt.figure()
    hist1 = hist1.item()
    hist2 = hist2.item()
    hist3 = hist3.item()
    train = hist1['loss'] + hist2['loss'] +hist3['loss']
    val   = hist1['val_loss'] + hist2['val_loss'] + hist3['val_loss']
    plt.plot(train)
    plt.plot(val)
    plt.title('Loss')
    plt.ylabel('loss')
    plt.xlabel('epoch')
    plt.legend(['Training', 'Validation'], loc='upper left')
    plt.show()

def plot_fine_hist2(hist, hist1, hist2, hist3):
    fig = plt.figure()
    hist = hist.item()
    hist1 = hist1.item()
    hist2 = hist2.item()
    hist3 = hist3.item()
    train = hist['loss'] + hist1['loss'] + hist2['loss'] +hist3['loss']
    val   = hist['val_loss'] + hist1['val_loss'] + hist2['val_loss'] + hist3['val_loss']
    plt.plot(train)
    plt.plot(val)
    plt.title('Loss')
    plt.ylabel('loss')
    plt.xlabel('epoch')
    plt.legend(['Training', 'Validation'], loc='upper right')
    plt.show()
    fig.savefig('plot.pdf')
def plot_fine_hist2(hist, hist1, hist2, hist3, hist4, hist5, hist6):
    fig = plt.figure()
    hist = hist.item()
    hist1 = hist1.item()
    hist2 = hist2.item()
    hist3 = hist3.item()
    hist4 = hist4.item()
    hist5 = hist5.item()
    hist6 = hist6.item()
    train = hist['loss'] + hist1['loss'] + hist2['loss'] +hist3['loss']+ hist4['loss'] + hist5['loss'] +hist6['loss']
    val   = hist['val_loss'] + hist1['val_loss'] + hist2['val_loss'] + hist3['val_loss']+ hist4['val_loss'] + hist5['val_loss'] + hist6['val_loss']
    plt.plot(train)
    plt.plot(val)
    plt.title('Loss')
    plt.ylabel('loss')
    plt.xlabel('epoch')
    plt.legend(['Training', 'Validation'], loc='upper right')
    plt.show()
    fig.savefig('plot.pdf')
```
